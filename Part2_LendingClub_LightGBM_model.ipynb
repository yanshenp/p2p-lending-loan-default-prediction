# Import the datasets


```python
import pandas as pd

train_x_fs = pd.read_csv('/kaggle/input/trial2-lendingclub-datasets/train_x_fs.csv')
test_x_fs = pd.read_csv('/kaggle/input/trial2-lendingclub-datasets/test_x_fs.csv')
train_y = pd.read_csv('/kaggle/input/trial2-lendingclub-datasets/train_y.csv')
test_y = pd.read_csv('/kaggle/input/trial2-lendingclub-datasets/test_y.csv')
```

# Function to evaluate model performance

```python
# Following codes will create a function to test the performance of models using the testing dataset
def model_performance(model_name, model_type, test_x, test_y):
    from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

    # Predict the y value using the model
    y_pred = model_name.predict(test_x)
    # Calculate evaluation metrics using metrics in sklearn
    accuracy = accuracy_score(test_y, y_pred)
    precision = precision_score(test_y, y_pred)
    recall = recall_score(test_y, y_pred)
    f1 = f1_score(test_y, y_pred)

    # Print the evaluation metrics
    print(f'-----{model_type}-----')
    print('Accuracy:', accuracy)
    print('Precision:', precision)
    print('Recall:', recall)
    print('F1 Score:', f1)
```

# Fine tune the LightGBM model using random search

```python
import lightgbm as lgb
from sklearn.model_selection import RandomizedSearchCV

# Define the parameter grid for random search
lgbm_param = {
    'num_leaves': list(range(20, 101, 1)),  # Range from 20 to 100 with step size 1
    'learning_rate': [0.01, 0.05, 0.1, 0.2],  # Specific values of learning rate
    'max_depth': list(range(5, 21, 5)) + [-1],  # Range from 5 to 20 with step size 5, including -1 for no limit
    'min_child_samples': list(range(10, 101, 10)),  # Range from 10 to 100 with step size 10
    'subsample': [0.6, 0.7, 0.8, 0.9, 1.0],  # Specific values of subsample ratio of training instances
    'colsample_bytree': [0.6, 0.7, 0.8, 0.9, 1.0],  # Specific values
    'boosting' : ['gbdt','dart', 'goss']
    # Add more parameters with ranges or specific values as needed
}

# First create a lgbm classifier
lgbm = lgb.LGBMClassifier(random_state=123)
# Perform RandomizedSearchCV to find the best combination of parameters
random_search = RandomizedSearchCV(estimator=lgbm, param_distributions=lgbm_param, 
                                   n_iter=100, scoring='accuracy', cv=5, verbose=0, n_jobs=-1, random_state=123)

random_search.fit(train_x_fs, train_y)

# Get the best parameters and best accuracy
lgbm_best_params = random_search.best_params_
best_accuracy = random_search.best_score_

print("Best Parameters:", lgbm_best_params)
print("Best Accuracy:", best_accuracy)
```

```python
# Export the random search results as a DataFrame
lightgbm_random_search_results_df = pd.DataFrame(random_search.cv_results_)

# Save the results to a CSV file
lightgbm_random_search_results_df.to_csv('lightgbm_random_search_results.csv', index=False)
```

```python
# Now, use the best combination of param to train a fine-tuned LightGBM model
lgbm = lgb.LGBMClassifier(**lgbm_best_params, random_state=123)
lgbm_finetuned = lgbm.fit(train_x_fs, train_y)
```

```python
# Save the model
import pickle
with open('/kaggle/working/lgbm_finetuned.pkl','wb') as f:
    pickle.dump(lgbm_finetuned,f)
```

```python
# Evaluate model performance
model_performance(lgbm_finetuned,'LightGBM fine tuned',test_x_fs, test_y)
```
